## Capping

lalu kita lakukan capping pada outlier kita, karena kalau kita trimming maka data tersebut akan habis

```python
column_selection = ['distance', 'surge_multiplier', 'price']

# Menghitung nilai kuartil
Q1 = df[column_selection].quantile(0.25)
Q3 = df[column_selection].quantile(0.75)

# Menghitung rentang interquartile (IQR)
IQR = Q3 - Q1

# Mendefinisikan batas atas dan batas bawah untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mengganti nilai outlier dengan batas atas atau batas bawah
df[column_selection] = df[column_selection].clip(lower_bound, upper_bound, axis=1)
df_capp =df[column_selection]
```

```python
df_capp.info()
```

Hasil data yang sudah di capping

```python
column_selection = ['distance', 'surge_multiplier', 'price']
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[column_selection])
plt.title('Boxplot of Selected Features')
plt.xticks(rotation=45)
plt.show()
```

Capping dilakukan dengan menggantikan nilai outlier yang di atas batas atas dengan batas atas itu sendiri, dan menggantikan nilai outlier yang di bawah batas bawah dengan batas bawah itu sendiri.
